### Prepocessing Data

```python
# melakukan preprocessing data
from sklearn.preprocessing import LabelEncoder
```

```python
# mengubah values kelas yang bertipe data  string/object menjadi numerik
for col in ['time_left', 'ct_score','map','bomb_planted', 'ct_health', 'ct_armor', 'ct_money', 'ct_helmets']:
        label_encoder = LabelEncoder()
        counter_strike[col] = label_encoder.fit_transform(counter_strike[col])
```

```python
counter_strike
```

```python
# memisaqhkan kolom target (round_winner) dari dataset
x = counter_strike.drop('round_winner', axis = 1)
y = counter_strike['round_winner']
```

```python
x
```

```python
y
```

```python
# spilling data training dan testing
from sklearn.model_selection import train_test_split
```

```python
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.20)
```

```python
x_train.shape
```

```python
x_test.shape
```

```python
y_train.shape
```

### Modelling

```python
from sklearn.naive_bayes import GaussianNB

# inisialisasi model
modelNB = GaussianNB()

#l latih model
modelNB.fit(x_train, y_train)
```

```python
hasil_prediksi = modelNB.predict(x_test)
```

```python
# import library
from sklearn.metrics import accuracy_score

# menghitung dan menampilkan akurasi model
print('Akurasi Model: ', accuracy_score(y_test, hasil_prediksi) * 100 )
```
